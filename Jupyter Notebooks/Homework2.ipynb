```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as mpl

%matplotlib inline
```

```python
df = pd.read_csv('Airbnb_Texas_Rentals.csv')
```

```python
df.columns
```

```python
# to prevent future UFuncTypeError: ufunc 'multiply' did not contain a loop with signature matching types (dtype(' dtype(', remove the following

illegal_entries = ['Studio']
df['average_rate_per_night'] = df.average_rate_per_night.map(lambda x: 0 if illegal_entries.__contains__(x) else float(x))
```

```python
df.info()
```

```python
# correlation analysis between columns average_rate_per_night and bedrooms_count with data types object. Need to convert object to float64.

df_avg_rpn_bcount = df[['average_rate_per_night', 'bedrooms_count']]
df_corr_ab = sns.regplot('average_rate_per_night', 'bedrooms_count', df_avg_rpn_bcount)
df_corr_ab.set(xlim = (0, 6000), ylim = (0, 12.5))
df_corr_ab.axhline(0, color = 'k', linestyle = '-', linewidth = 2)
df_corr_ab.avxline(0, color = 'k', linestyle = '-', linewidth = 2)
df_corr_ab.axes.set_title('Correlation: avg. rate per night & bedrooms count')
mpl.show()

for entry in df_avg_rpn_bcount.columns:
    print("Column: {}".format(entry))
    data = df_avg_rpn_bcount.loc[:,entry]
    data.hist()
    print("- mean: {}".format(data.mean()))
    print("- std: {}".format(data.std()))
    mpl.show()
```
