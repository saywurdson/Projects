```python
import requests
from bs4 import BeautifulSoup
import datetime
```

```python
#Gets ASHP revisions drug shortage information
r = requests.get('https://www.ashp.org/Drug-Shortages/Current-Shortages/Drug-Shortages-List?page=All&sort=2')
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
#Sets date time to grab for revision data on ASHP site, set to yesterday right now
re_date = datetime.datetime.now() - datetime.timedelta(days=1)
```

```python
#Parses ASHP landing paging to find drug shortage row information
soup_tr = soup.find('tbody').find_all('tr', limit=30)

page_list = []
for item in soup_tr:
    l = item.find_all('td')
    if l[2].text == re_date.strftime(' %b %d, %Y'):
        page_list.append(l[0].find('a')['href'])
```

```python
#loops over pages that were updated on date of interest and grabs the corrisponding shortage information page
for page in page_list:
    r = requests.get('https://www.ashp.org/Drug-Shortages/Current-Shortages/' + page)
    soup = BeautifulSoup(r.content, 'html.parser')
```

```python
#using shortage information page this will grap all <li> text under "affected" heading
affected = soup.find(id='1_Affected').find_all('li')
for item in affected:
    #need to figure out what you want to do with this and extract NDC
    print(item.text)

#using shortage information page this will grap all <li> text under "affected" heading
available = soup.find(id='1_Avaliable').find_all('li')
for item in available:
    #need to figure out what you want to do with this and extract NDC
    print(item.text)
```
